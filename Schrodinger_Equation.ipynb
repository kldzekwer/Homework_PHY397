Schrodingers Equation

```python
#defining constants
import numpy as np
import math
import matplotlib.pyplot as plt
x_min= -6
x_max = 6
N = 1000 #Number of grid points and if m and \hbar is assumed to be 1
d = np.linspace(x_min, x_max, N)
h = d[1]-d[0] #Step size
K_E = np.zeros((N-2, N-2))
for i in range (N-3):
  K_E[i,i]-=2
  K_E[i, i+1]+=1
  K_E[i+1, i]+=1
print(K_E)
P_E = np.zeros((N-2, N-2))
for i in range(N-2):
  P_E[i,i] = d[i]**2
print(P_E)
H = -1/(2*h**2)*(K_E+P_E) #creating hamiltonian matrix
print(H)
val,vec=np.linalg.eig(H) #finding the eigen value of the hamiltonian
z = np.argsort(val)
z = z[0:4]
#The eigen values are
energies=(val[z]/val[z][0])
print("The eigen values are ",val)
print("The eigen functions are",vec)
for i in range(len(z)):
    y = []
    y = np.append(y,vec[:,z[i]])
    y = np.append(y,0)
    y = np.insert(y,0,0)
    plt.plot(d,y)
    plt.xlabel('x', size=14)
    plt.ylabel('$\psi$(x)',size=14)
plt.legend()
plt.title('normalized wavefunctions for a harmonic oscillator using finite difference method',size=14)
plt.show()
#plt.plot(d, z)


```
